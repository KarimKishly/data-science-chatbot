```python
import pandas as pd
import numpy as np
import json
```

```python
data1 = pd.read_csv('../assets/dataset_v1.csv')
data1
```

```python
data2 = pd.read_csv('../assets/dataset_v2.csv')
data2
```

```python
data3 = pd.read_csv('../assets/dataset_v3.csv')
data3
```

```python
patient_confirmation = data2['Patient confirms the appointment'].values.tolist()
patient_rejection = data2['Patient rejects the appointment'].values.tolist()
chatbot_confirmation1 = data2['Chatbot confirms the appointment'].values.tolist()
chatbot_rejection_response1 = data2['Chatbot responds to rejection'].values.tolist()

with open('../assets/patient/patient_confirmation.txt', 'w') as f:
    f.write(json.dumps(patient_confirmation))
with open('../assets/patient/patient_rejection.txt', 'w') as f:
    f.write(json.dumps(patient_rejection))
with open('../assets/chatbot/bot_confirmation.txt', 'w') as f:
    f.write(json.dumps(chatbot_confirmation1))
with open('../assets/chatbot/bot_rejection_response.txt', 'w') as f:
    f.write(json.dumps(chatbot_rejection_response1))
```

```python
with open('../assets/patient/patient_greetings.txt', 'r') as f:
    patient_greetings = json.loads(f.read())
with open('../assets/patient/patient_issue.txt', 'r') as f:
    patient_issue = json.loads(f.read())
with open('../assets/patient/patient_confirmation.txt', 'r') as f:
    patient_confirmation = json.loads(f.read())
with open('../assets/patient/patient_rejection.txt', 'r') as f:
    patient_rejection = json.loads(f.read())

greetings = np.concatenate([
    patient_greetings,
    patient_issue,
    patient_confirmation,
    patient_rejection]
)
message_type = (
        ['patient_greet'] * len(patient_greetings)
        + ['patient_issue'] * len(patient_issue)
        + ['patient_confirmation'] * len(patient_confirmation)
        + ['patient_rejection'] * len(patient_rejection)
)

data_dict = {
    'message': greetings,
    'message_type': message_type
}
df = pd.DataFrame(data_dict)
df.to_csv('../assets/patient/patient_concat.csv')
df
```
