```python
patient_confirmation = data2['Patient confirms the appointment'].values.tolist()
patient_rejection1 = data2['Patient rejects the appointment'].values.tolist()
chatbot_confirmation1 = data2['Chatbot confirms the appointment'].values.tolist()
chatbot_rejection_response1 = data2['Chatbot responds to rejection'].values.tolist()

with open('assets/patient/patient_confirmation.txt', 'w') as f:
    f.write(json.dumps(patient_confirmation))
with open('assets/patient/patient_rejection.txt', 'w') as f:
    f.write(json.dumps(patient_rejection1))
with open('assets/chatbot/bot_confirmation.txt', 'w') as f:
    f.write(json.dumps(chatbot_confirmation1))
with open('assets/chatbot/bot_rejection_response.txt', 'w') as f:
    f.write(json.dumps(chatbot_rejection_response1))
```

```python
with open('assets/patient/patient_greetings.txt', 'r') as f:
    patient_greetings = json.loads(f.read())
with open('assets/patient/patient_issue.txt', 'r') as f:
    patient_issue = json.loads(f.read())
with open('assets/patient/patient_confirmation.txt', 'r') as f:
    patient_confirmation = json.loads(f.read())
with open('assets/patient/patient_rejection.txt', 'r') as f:
    patient_rejection1 = json.loads(f.read())

greetings = np.concatenate([
    patient_greetings,
    patient_issue,
    patient_confirmation,
    patient_rejection1]
)
message_type = (
        ['patient_greet'] * len(patient_greetings)
        + ['patient_issue'] * len(patient_issue)
        + ['patient_confirmation'] * len(patient_confirmation)
        + ['patient_rejection'] * len(patient_rejection1)
)

data_dict = {
    'message': greetings,
    'message_type': message_type
}
df = pd.DataFrame(data_dict)
df.to_csv('assets/patient/patient_concat.csv')
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import joblib

tfidf_vectorizer = TfidfVectorizer(lowercase=True)
X_full = df['message'].values
y_full = df['message_type'].values
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['message_type'], test_size=0.2)
message_tfidf = tfidf_vectorizer.fit_transform(X_train)

nb_classifier = MultinomialNB()
nb_classifier.fit(message_tfidf, y_train)

svm = SVC()
svm.fit(message_tfidf, y_train)

print(nb_classifier.score(tfidf_vectorizer.transform(X_test), y_test))
print(svm.score(tfidf_vectorizer.transform(X_test), y_test))

message_tfidf = tfidf_vectorizer.fit_transform(X_full)
svm.fit(message_tfidf, y_full)   
nb_classifier.fit(message_tfidf, y_full)

joblib.dump(tfidf_vectorizer, 'assets/vectorizer.joblib')   
joblib.dump(nb_classifier, 'assets/nb_model.joblib')
joblib.dump(svm, 'assets/svm_model.joblib')

# test_message = 'I am feeling down'
# test_tfidf = tfidf_vectorizer.transform(np.array([test_message]))
# 
# prediction = nb_classifier.predict(test_tfidf)
# print(prediction)
```
